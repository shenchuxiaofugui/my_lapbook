```python

import os
import pandas as pd
import numpy as np
```

```python
#整理医生给的数据表
#先剔除没有文件的数据
#先把有文件的名字做成表格，再跟取交集
path=r"\\mega\syli\dataset\Primary and metastatic\Primary"
biao_path=r"\\mega\syli\dataset\Primary and metastatic\primary.csv"
files=os.listdir(path)
file_name=pd.DataFrame(files,columns=['MRI-id'])
df=pd.read_csv(biao_path)
df1=pd.merge(file_name,df)
df1=df1.drop(labels=[df1.columns[1],'impatient-id',df1.columns[-1]],axis=1)  #删除并重新赋值，df1.columns[1]是第2列的列名
#df1.to_csv(r"\\mega\syli\dataset\Primary and metastatic\metastatic1.csv",index=False,encoding="utf_8_sig")   
#后面参数是取消索引的意思，用sig就不会有乱码了。。
```

```python
#均值填充
for column in list(df1.columns[df1.isnull().sum() > 0]):
    print(column)
    df1[column]=df1[column].astype('float64')   #消掉字符串的格式，以便下面mean函数的计算
    mean_val = round(df1[column].mean())
    df1[column].fillna(mean_val, inplace=True)
df1.to_csv(r"\\mega\syli\dataset\Primary and metastatic\primary1.csv",index=False,encoding="utf_8_sig")
```

```python
biao_path=r"\\mega\syli\dataset\Primary and metastatic"
label1_path=biao_path+'\\primary1.csv'
label1=pd.read_csv(label1_path)
label1.insert(1,'label',1)
label0_path=biao_path+'\\metastatic1.csv'
label0=pd.read_csv(label0_path,converters={0:str})  #保留开头的0
label0.insert(1,'label',0)
df1=pd.concat([label0,label1])
df1.to_csv(biao_path+'\\caseinf.csv',index=False,encoding="utf_8_sig")
```
