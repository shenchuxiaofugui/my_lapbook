```python
import os
import pandas as pd
import SimpleITK as sitk
```

```python
#输出文件夹中图片信息
path=r'\\219.228.149.7\syli\dataset\Primary and metastatic\zheci\Primary and metastatic OC\Primary and metastatic OC\all'
dirs = os.listdir( path )
df = pd.DataFrame(columns = ['casename','spacing','direction','origin','size','real'] ) #columns = ['spacing','direction','origin','size','real'] 
# 输出所有文件和文件夹
for file in dirs:
    file1=path+'\\'+file+'\\t1.nii'
    if os.path.exists(file1):
        img=sitk.ReadImage(file1)
    else:
        continue
    b=sitk.GetArrayFromImage(img)
    spacing = img.GetSpacing()
    direction = img.GetDirection()
    origin = img.GetOrigin()
    size = img.GetSize()
    c={'casename':file,'spacing':spacing,'direction':direction,'origin':origin,'size':size,'real':b.shape}
    #print(c)
    df=df.append(c,ignore_index=True)  #必须要有返回值！！！！
    #print(spacing,direction,origin,size,b.shape)
df.to_csv(r'C:\Users\Administrator\Desktop\ovarioncus\xinxi.csv', index=False)
```
